```python
# Crawl Data

filename = 'motor_listrik.csv'
search_keyword = 'motor listrik lang:id until:2024-07-01 since:2020-01-01'
limit = 5000

# Run tweet-harvest
!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(filename, delimiter=",")

# Check the column names
print(df.columns)

# Define cleansing function
def cleanse_tweet(tweet):
    # Remove URLs
    tweet = re.sub(r'https?:\/\/\S+', '', tweet)
    # Remove usernames
    tweet = re.sub(r'@[\w\-]+', '', tweet)
    # Remove special characters
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)
    # Convert to lowercase
    tweet = tweet.lower()
    return tweet

# Determine the correct column for tweet text
text_column = 'content'  # Assuming 'content' is the correct column, adjust if necessary

# Apply cleansing function to the text column
if text_column in df.columns:
    df[text_column] = df[text_column].apply(lambda x: cleanse_tweet(x))
else:
    raise KeyError(f"The expected text column '{text_column}' was not found.")

# Remove duplicates
df.drop_duplicates(subset=text_column, keep='first', inplace=True)

# Display the DataFrame
display(df)

# Check the number of tweets
num_tweets = len(df)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```
