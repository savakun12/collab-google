```python
# Crawl Data

filename = 'motor_listrik.csv'
search_keyword = 'motor listrik lang:id until:2024-07-01 since:2020-01-01'
limit = 5000

!npx -y tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" --tab "LATEST" -l {limit} --token {twitter_auth_token}

import pandas as pd
import re
```

```python
# Specify the path to your CSV file
file_path = filename  # "tweets-data/" prefix not necessary if file is in the same directory

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
display(df)

# Cek jumlah data yang didapatkan
num_tweets = len(df)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```

```python
# Pembersihan Data

# Menghapus baris yang tidak diperlukan
df = df.dropna(subset=['created_at', 'user_screen_name', 'text'])

# Mengubah format tanggal
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Membersihkan nama pengguna
df['user_screen_name'] = df['user_screen_name'].str.strip().str.lower()
```

```python
# Membersihkan isi tweet
def bersihkan_tweet(tweet):
    tweet = re.sub(r'http\S+', '', tweet)  # Menghapus URL
    tweet = re.sub(r'@\w+', '', tweet)     # Menghapus mention
    tweet = re.sub(r'#\w+', '', tweet)     # Menghapus hashtag
    tweet = re.sub(r'[^\w\s]', '', tweet)  # Menghapus karakter khusus
    tweet = tweet.lower()                  # Mengubah ke huruf kecil
    return tweet

df['text'] = df['text'].apply(bersihkan_tweet)
```

```python
# Mengatasi duplikasi
df = df.drop_duplicates()

# Mengganti nama kolom sesuai keinginan
df = df.rename(columns={'created_at': 'tanggal', 'user_screen_name': 'nama_user', 'text': 'isi_tweet'})

# Memilih kolom yang relevan
df = df[['tanggal', 'nama_user', 'isi_tweet']]

# Menyimpan data yang telah dibersihkan ke dalam file CSV
df.to_csv('motor_listrik_bersih.csv', index=False)

# Display cleaned DataFrame
display(df)
print(f"Data yang telah dibersihkan disimpan dalam 'motor_listrik_bersih.csv'.")
```
